### 4:Churn Analysis

```python
# Starting Churn Analysis its very important to check which user is using our service or not 

plt.figure(figsize=(8,5))
sns.countplot(x="churned",data=df)
plt.title("Customer or churn distribution")
plt.show()

# its checking for how many user are active
(df['churned'] == 'Active').sum()
```

```python
# checking oder frequency of active user or non active user 

# code of plot
plt.figure(figsize=(8,5))
sns.boxplot(x='churned', y='order_frequency', data=df)
plt.title('Order Frequency or Churn')
plt.show()
```

```python
# Checking loyalty points of active or non active user

plt.figure(figsize=(8,5))
sns.boxplot(x='churned', y='loyalty_points', data=df)
plt.title('Loyalty Points or Churn')
plt.show()
```
